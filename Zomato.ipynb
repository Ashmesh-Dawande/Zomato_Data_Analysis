We will drop Best Seller column because the missing values is more than 50 percent.

```python
data = data.drop(data.columns[9] , axis=1)
data
```

```python
mean_value = data['Delivery Rating'].mean()
data['Delivary rating'] = data['Delivery Rating'].fillna(mean_value)
```

```python
new_mean = data['Dining Rating'].mean()
data['Dining rating'] = data['Dining Rating'].fillna(new_mean)
```

```python
data
```

As the columns containing missing values are numerical data and that too both are ratings so we will fill the na values with the mean of the columns 

```python
data = data.drop(['Dining Rating','Delivery Rating'],axis=1)
```
